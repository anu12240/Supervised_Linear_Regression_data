```python
#Import Pandas as pd
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
# Import Dataset
ds = pd.read_csv("salary_Data.csv")
ds
```

```python
ds.columns
```

```python
# Dividing Dataset as independent (input) and dependent(output) variables
x = ds [['YearsExperience']]
y = ds [['Salary']]
```

```python
# Plot Graph 
plt.scatter(x,y)
plt.show()
```

```python
x
```

```python
y
```

```python
# Spliting dataset as Traning and Testing
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.2, random_state = 0)
```

```python
# Introduce model
from sklearn.linear_model import LinearRegression
reg = LinearRegression()
reg.fit(x_train,y_train)
```

```python
# Predict
result = reg.predict([[10]])
result
```
